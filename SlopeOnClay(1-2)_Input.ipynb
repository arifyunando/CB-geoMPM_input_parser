```python
import mp_parser
import matplotlib.pyplot as plt


# Mesh Generation
a = mp_parser.mesh(.5)
a.setMesh(40, 15)
filled = a.nodes
# print(filled)
x = [i[0] for i in filled]
y = [i[1] for i in filled]
plt.plot(x,y, '.')
plt.axis('scaled')
plt.show()

a.printFile(digits = 3, folderName = "Slope1-2")
```

```python
points = [
    (0  , 0),
    (39.875, 0),
    (39.875, 2.875),
    (0  , 2.875)
]

points2 = [
    (0  , 0),
    (15.875, 0),
    (7.875, 3.875),
    (0  , 3.875)
]

# Bottom Layer [Clay]
b = mp_parser.particle(points, delta = 0.125)
b.translate(0.0625, 0.0625)

# Top Layer [Peat]
c = mp_parser.particle(points, delta = 0.125)
c.translate(0.0625, 3.0625)

# Fill [Sand]
d = mp_parser.particle(points2, delta = 0.125)
d.translate(0.0625, 6.0625)

filled1 = b.nodes
x1 = [i[0] for i in filled1]
y1 = [i[1] for i in filled1]

filled2 = c.nodes
x2 = [i[0] for i in filled2]
y2 = [i[1] for i in filled2]

filled3 = d.nodes
x3 = [i[0] for i in filled3]
y3 = [i[1] for i in filled3]

plt.plot(x1,y1, '.')
plt.plot(x2,y2, '.')
plt.plot(x3,y3, '.')

plt.axis('scaled')
plt.show()

b.printFile(fileName = 'clay', folderName = "Slope1-2", digits=4)
c.printFile(fileName = 'peat', folderName = "Slope1-2", digits=4)
d.printFile(fileName = 'sand', folderName = "Slope1-2", digits=4)
```

```python
# Side Boundaries
nset = []
for key, item in enumerate(a.nodes):
    if item[0] == 0 or item[0] == 40:
#         print(key, item)
        nset.append(key)
print(nset)
```

```python
# Bottom Boundaries
nset = []
for key, item in enumerate(a.nodes):
    if item[1] == 0:
#         print(key, item)
        nset.append(key)
print(nset)
```

```python
# Particle Loading
particles = b.nodes + c.nodes + d.nodes
pset = []
tmp = []
for key, item in enumerate(particles):
    if item[1] > 6:
#         print(key, item)
        tmp.append(item[1])
tmp = set(tmp)
tmp = list(tmp)

nodes = []
for y in tmp:
    row = []
    for key, item in enumerate(particles):
        if item[1] == y:
            row.append(item[0])
    nodes.append([max(row), y])
    
for key, item in enumerate(particles):
    if item[1] > 9.93:
        pset.append(key)
    if item in nodes:
        pset.append(key)
        
pset = set(pset)
pset = list(pset)
pset.sort()
print(len(particles))
for i in pset:
    print(particles[i])
print(pset)
```
